```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Souvik Bhattacharyya\Desktop\Mentorship\iris_dataset.txt", header=None)
df.head()
```

```python
data = df.iloc[:, :-1]
label = df.iloc[:, -1]
```

```python
# Splitting the dataset into test and train data
from sklearn.model_selection import train_test_split
x = data
y = label
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Linear Kernel
model_1 = OneVsRestClassifier(SVC(kernel='linear', decision_function_shape='ovo'))
model_1.fit(x_train, y_train)
prediction_1 = model_1.predict(x_test)
print(f"Accuracy Using Linear Kernel: {accuracy_score(y_test, prediction_1) * 100} % ")    

# Ploynomial Kernel
model_2 = OneVsRestClassifier(SVC(kernel='poly', decision_function_shape='ovo'))
model_2.fit(x_train, y_train)
prediction_2 = model_2.predict(x_test)
print(f"Accuracy Using Polynimial Kernel: {accuracy_score(y_test, prediction_2) * 100} % ")    

# RBF Kernel
model_3 = OneVsRestClassifier(SVC(kernel='rbf', decision_function_shape='ovo'))
model_3.fit(x_train, y_train)
prediction_3 = model_3.predict(x_test)
print(f"Accuracy Using RBF Kernel: {accuracy_score(y_test, prediction_3) * 100} % ")    
```
